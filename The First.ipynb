```python
'''importing libraries'''

import pandas as pd
import matplotlib.pyplot as plt
import os


'''LOAD DATA SETS FROM KAGGLEHUB'''
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shivamb/netflix-shows")

print("Path to dataset files:", path)

# Correcting the file path to read the CSV file
df = pd.read_csv(os.path.join(path, "netflix_titles.csv"))
df.head()
df.shape
df.isnull().sum()
df.dtypes
df.head()

type_counts = df['type'].value_counts()
print(type_counts)

type_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Movies vs TV Shows on Netflix")
plt.ylabel("")
plt.show()

df['genres'] = df['listed_in'].fillna("Unknown") # Assuming 'listed_in' column represents genres

top_genres = df['genres'].value_counts().head(10)
print(top_genres)

top_genres.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Most Common Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

df['director'] = df['director'].fillna("Unknown")

top_directors = df['director'].value_counts().head(10)
print(top_directors)

top_directors.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Directors")
plt.xlabel("Director")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

print("INSIGHTS:")

print("\n1. Netflix contains more Movies than TV Shows.")
print("2. Drama, Comedy, and Action appear as the most common genres.")
print("3. The United States produces the highest number of Netflix titles.")
print("4. Some directors appear very frequently, showing strong partnerships with Netflix.")
print("5. Data shows Netflix content is diverse and spans many countries/genres.")
```
